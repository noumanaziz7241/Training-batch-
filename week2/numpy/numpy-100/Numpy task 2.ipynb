Numpy Practice Problem Crop and Rotate Image using Numpy


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


im_path = "baboon.png"
im_array = cv2.imread(im_path)
print("original")
plt.imshow(cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB))
plt.show()


# this image has 512 pixels width and 512 pixels height
# print(im_array.shape)


```

```python
im_array.shape
```

```python
# crop the image using numpy 10 percent from top, 20 percent from bottom
# 5 percent from left and 25 percent from right
height,width= im_array.shape[0],im_array.shape[1]
top, bottom, left, right= 0.1 ,0.2 , 0.05, 0.25
cropped_image= im_array[0+int(top*height):height-int(height*bottom),0+ int(width*left):width-int(width*right)]
cropped_image.shape
```

```python
# cropped_im_array = your code here
print("cropped")
plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
plt.show()


```

```python
degrees = 90
rotated_im_array = im_array.copy()
for _ in range(degrees // 90):
    rotated_im_array = rotated_im_array.transpose(1, 0, 2)
    for j in range(0, rotated_im_array.shape[1] // 2):
        c = rotated_im_array[:, j, :].copy()
        rotated_im_array[:, j, :] = rotated_im_array[: , rotated_im_array.shape[1]-j-1, :]
        rotated_im_array[: , rotated_im_array.shape[1]-j-1, :] = c
print("rotated")
plt.imshow(cv2.cvtColor(rotated_im_array, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
rotated_im_array = np.rot90(im_array,k=3)
print("rotated")
plt.imshow(cv2.cvtColor(rotated_im_array, cv2.COLOR_BGR2RGB))
plt.show()
```
