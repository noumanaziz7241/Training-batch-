```python
import json
import matplotlib.pyplot as plt
# Opening JSON file
f = open('/home/nouman/Downloads/AnomalyEventDetection/Anomaly Event Detection/JSON/hist_exp.json')
  
# returns JSON object as 
# a dictionary
history= json.load(f)
plt.rcParams["figure.figsize"] = (10,8)
loss = history['loss']
val_loss = history['val_loss']
epochs = range(1, len(loss) + 1)
#epochs= range(1,len(history['loss']+1))
plt.plot(epochs, loss, label='Training Loss')
plt.plot(epochs,val_loss,color='g', label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss') 
#plt.ylim(0.2,1.0)
#plt.xlim(0,len(data1['normality_score'][1439:2651]))
plt.legend(["Training Loss", "Validation Loss"], loc ="best")
plt.title("Training and Validation Loss")
plt.show() 
# Closing file
f.close()
```

```python
import json
import matplotlib.pyplot as plt
# Opening JSON file
f = open('/home/nouman/Downloads/AnomalyEventDetection/Anomaly Event Detection/JSON/hist_exp.json')
  
# returns JSON object as 
# a dictionary
history= json.load(f)
plt.rcParams["figure.figsize"] = (10,8)
acc = history['accuracy']
val_acc = history['val_accuracy']

epochs = range(1, len(acc) + 1)
plt.plot(epochs, acc, label='Training Accuracy')
plt.plot(epochs,val_acc,color='g', label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy') 
plt.legend(["Training Accuracy", "Validation Accuracy"], loc ="best")
plt.title("Training and Validation Accuracy")
plt.show() 
# Closing file
f.close()
```
